## **Función para determinar el Porcentaje de nulos de un Dataset**

```python
def nulos(df):
  cant = len(df)
  for col in df.columns:
    print(col,':',round(df[col].isnull().sum() * 100 / cant,2),'%')

nulos(df)
```
