```r
#Cargamos la data
Datos<-iris
head(Datos)
```

```r
#Instalamos el paquetee de ggplot2 para Visualizacion de Datos
install.packages("ggplot2")
```

```r
#ScatterPlot con ggplot
library(ggplot2)
ggplot(data=Datos)+
  geom_point(aes(x=Sepal.Length,
                 y=Sepal.Width))
```

```r
#Aumentar el tamaño de los puntos
ggplot(data=Datos)+
  geom_point(aes(x=Sepal.Length,
                 y=Sepal.Width),
                 size=3)
```

```r
#Cambiar el color de los datos
ggplot(data=Datos)+
  geom_point(aes(x=Sepal.Length,
                 y=Sepal.Width),
                 size=3,
                 color="blue")
```

```r
#Utilizar el color pero por una variable - Species
ggplot(data=Datos)+
  geom_point(aes(x=Sepal.Length,
                 y=Sepal.Width,
                 color=Species),
                 size=3)
```

```r
#Darle un poco de transparencia a los puntos (este parametro alpha varia entre 0 y 1)
ggplot(data=Datos)+
  geom_point(aes(x=Sepal.Length,
                 y=Sepal.Width,
                 color=Species),
                 alpha = .4,
                 size=3)
```

```r
#Podemos usar una variable categórica para q haga un grafico para cada una de sus categorias
ggplot(data=Datos)+
  geom_point(aes(x=Sepal.Length,
                 y=Sepal.Width),
                 alpha = .4,
                 size=3)+
  facet_wrap(~Species)
```
