```python
#Carga de las librerías
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

import xgboost as xgb #pip install xgboost
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings("ignore")
```

```python
# Leer el dataset
url = 'https://raw.githubusercontent.com/Geerdata/DS/main/Datacoder/Arch.%20Modelo/WA_Fn-UseC_-Telco-Customer-Churn.csv'
data= pd.read_csv(url,sep=',')
data.head()
```

```python
#Separamos en X e y
X = data[['InternetService', 'TotalCharges', 'Contract', 'PaymentMethod']].copy()
y = data["Churn"]
```

```python
x = pd.get_dummies(X)
```

```python
#Separamos en train y test!
(X_train, X_test,y_train, y_test) = train_test_split(x,y,stratify=y,test_size=0.30,random_state=42)
```

```python
clf_xgb = xgb.XGBClassifier(n_estimators=10,seed=42,max_depth=6, learning_rate=0.01)
```

```python
X_test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test  = le.fit_transform(y_test)
```

```python
clf_xgb.fit(X_train,y_train) #Entrenamos el modelo
```

```python
y_train_pred = clf_xgb.predict(X_train) #Prediccion en Train
y_test_pred = clf_xgb.predict(X_test) #Prediccion en Test
```

```python
from sklearn.metrics import classification_report
reporte=classification_report(y_test,y_test_pred)
print(reporte)
```
