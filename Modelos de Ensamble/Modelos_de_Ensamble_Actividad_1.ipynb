#**Adaboost**

Como no podía ser de otra manera, vamos a ver una implementación del algoritmo de Adaboost en nuestro famoso dataset de Iris :)

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
#Cargamos la data!
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
#Separación en Train y Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
# Creamos un AdaBoostClassifier
ada = AdaBoostClassifier(n_estimators=50,learning_rate=0.1)
```

```python
# Chequear parametros
#AdaBoostClassifier?
```

```python
#Fiteamos el modelo
model = ada.fit(X_train, y_train)
```

```python
#Predicción!
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#Accuracy!
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

#**Gradient Boosting**

```python
#Importamos las librerias
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
import pandas as pd
X = pd.DataFrame(cancer.data, columns = cancer["feature_names"])
y = cancer["target"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
#Veamos un simple arbol de decision!
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
tree.score(X_test, y_test)
```

```python
#Ahora comparemos con GBM
from sklearn.ensemble import GradientBoostingClassifier
gbrt = GradientBoostingClassifier(random_state = 42)
gbrt.fit(X_train, y_train)
gbrt.score(X_test, y_test)
```

```python
import xgboost as xgb #pip install xgboost
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split
```

```python
url = 'https://raw.githubusercontent.com/Geerdata/DS/main/Datacoder/CSV/winequality-red.csv'
data = pd.read_csv(url, sep=',')
data.head()
```

```python
data['quality'].unique()
```

Vamos a hacer un problema de clasificacion. Asi que lo que vamos a predecir es
si el vino es de buena calidad o no. Tomaremos el 6 como umbral para decidir si
es bueno o no.

```python
data.loc[data['quality'] < 6, 'quality'] = 0 #baja calidad
data.loc[data['quality'] >= 6, 'quality'] = 1 #alta calidad
```

```python
#Veamos que tenemos!
data
```

```python
X = data.drop("quality", axis=1) #Elimino de mi dataset la variable a predecir
y = data.quality #Defino el Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=123)
```

```python
clf_xgb = xgb.XGBClassifier(objective='binary:logistic', n_estimators=10,seed=42,max_depth=6, learning_rate=0.01)
```

```python
clf_xgb.fit(X_train,y_train) #Entrenamos el modelo
```

```python
y_train_pred = clf_xgb.predict(X_train) #Prediccion en Train
y_test_pred = clf_xgb.predict(X_test) #Prediccion en Test
```

```python
from sklearn.metrics import accuracy_score

#Calculo el accuracy en Train
#train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

#print('% de aciertos sobre el set de entrenamiento:', train_accuracy)
print('% de aciertos sobre el set de evaluación:',test_accuracy)
```
