# **Group by**

```python
df = pd.DataFrame({'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5),
                   'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one']})
df
```

```python
grouped =df['data1'].groupby(df['key1'])
```

```python
grouped
```

```python
grouped.median()
```

```python
df['data1'].groupby(df['key1']).sum()
```

```python
df['data1'].groupby([df['key1'],df['key2']]).sum()
```
