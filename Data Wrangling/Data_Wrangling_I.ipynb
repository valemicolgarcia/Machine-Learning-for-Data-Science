# MERGE Y CONCAT

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'first': range(7)})
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'second': range(3)})
```

La variable key que unifica debe estar en ambos dataframes

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on ='key',how= 'inner') #intersaccion de los datos
```

```python
pd.merge(left=df1,right=df2, on ='key',how='left') #me traigo la intersaccion entre ambos df + datos que se encontraban solo en la tabla a
```

```python
pd.merge(df1,df2, on ='key',how= 'outer') #me trae por completo toda la informacion
```

```python
pd.merge(df1,df2, on ='key',how= 'left')
```

```python
pd.merge(df1,df2, on ='key',how= 'right')
```
