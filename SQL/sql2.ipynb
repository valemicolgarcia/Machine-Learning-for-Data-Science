# **CONEXION (Acceso de BBDD con SQL Lite)**

Para ejecutar consultas en SQLite, debemos usar ```cursor.execute()```. Definiremos una funci√≥n llamada ```execute_query()``` que usa este m√©todo. Su funci√≥n aceptar√° el objeto de conexi√≥n y una cadena de consulta, que pasar√° a ```cursor.execute()```.

```.execute()``` puede ejecutar cualquier consulta que se le pase en forma de string.


````python
import sqlite3
import pandas as pd
from sqlite3 import Error # Importar librerias
````

````python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query ejecutado satisfactoriamente")
    except Error as e:
        print(f"El error '{e}' ha ocurrido")
````

````python
#Create una conexion de lectura
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"Error '{e}' ha ocurrido")
````

````python
elegir_category = "SELECT * from category"
categorys = execute_read_query(sql, elegir_category)

for categ in categorys:
  print(categ)
````

````python
elegir_film = "SELECT * from film_category"
films = execute_read_query(sql, elegir_film)

for film in films:
  print(film)
````

````python
joins_category_film = """
SELECT
  *
FROM
  category
  INNER JOIN film_category ON category.category_id = film_category.category_id
"""
category_film = execute_read_query(sql, joins_category_film)

for x in category_film:
    print(x)
````

````python
joins_film_category = """
SELECT
  *
FROM
  category
  INNER JOIN film_category ON category.category_id = film_category.category_id
  INNER JOIN film ON film.film_id = film_category.film_id
"""
film_film = execute_read_query(sql, joins_film_category)

for r in film_film:
    print(r)
````
