```python
df.Contract.unique()
```

```python
x = df[['InternetService', 'TotalCharges', 'Contract', 'PaymentMethod']].copy()
```

```python
x
```

```python
y =df[['Churn']].copy()
```

```python
y
```

```python
print(x.InternetService.unique())
print(x.Contract.unique())
print(x.PaymentMethod.unique())
```

```python
x_new = pd.get_dummies(x)
```

```python
x_new.shape
```

```python
x_new.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_new, y, test_size=0.33, stratify= y, random_state=42)
```

```python
# Entrenar el arbol
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
y_pred= clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true=y_test,y_pred=y_pred))
```
