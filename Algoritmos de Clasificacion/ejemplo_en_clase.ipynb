```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
url='https://raw.githubusercontent.com/Geerdata/DS/main/Datacoder/Arc.%20Modelo/WA_Fn-UseC_-Telco-Customer-Churn.csv'

df= pd.read_csv(url,sep=',')

df.head()
```

```python
x = df[['InternetService', 'TotalCharges', 'Contract', 'PaymentMethod']].copy()
y = df[['Churn']].copy()
```

```python
x.head()
```

```python
x_encoded = pd.get_dummies(x)
x_encoded.head()
```

Ahora no va 0,1 sino que muestra true false

```python
x_encoded.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.3, random_state=42)#test depende del tamaÃ±o del dataset en general, semilla de donde arranca
```

```python
# Entrenar el arbol
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
pred=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true=y_test,y_pred=pred))
```

De todo el muestreo se llega al 77% de acertividad, a nivel entrenamiento es bueno
- Precision: El objetivo principal es predecir los que se van a dar de baja, nuestro modelo tiene mas precision en las personas que no se van a dar de baja, con respecto a las que si se van a dar de baja.
